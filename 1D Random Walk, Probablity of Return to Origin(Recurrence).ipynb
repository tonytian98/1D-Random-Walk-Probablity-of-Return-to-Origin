```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# original position
o=0

# different choice of step length
step1=1
step2=-1

#Number of different probabilities wanted
nump=5
# probability vector of taking step1
P=np.linspace(0,1,nump)

# Number of iteration
numit=100
```

```python
#walk n times, if returned to origin ,return 1; if not after n steps, return 0
def WalkNTimes(n,p):
    step=0
    position=o
    while True:
        position+=np.random.choice([step1,step2],p=[p,1-p])
        step+=1
        if position==o:
            return 1
            
        if step==n:
            return 0
            
```

```python
#return the rate of recurrence(back to origin) when the probability of taking step1 is p, step2 is 1-p
def WalkWithP(p):
    POfRec=[]
    for n in range(1,201,1):
        Recurrent=[]
        for i in range(numit):
            Recurrent.append(WalkNTimes(n,p))
        POfRec.append(np.mean(Recurrent))
    return POfRec
```

```python
trackOfP=[]
for p in P:
    trackOfP.append(WalkWithP(p))
```

```python
for i in range(nump):
    plt.plot(np.linspace(1,200,200),trackOfP[i])

le=[]
for i in range(nump):    
    le.append('p={}'.format(P[i]))    
plt.legend(le,loc='best')

plt.xlabel('Number of steps in one walk')
plt.ylabel('rate of recurrence \n (in {} tries)'.format(numit))
```

```python
# actually in one dimensional random walk
# the probability of recurrence is 1 iff p=0.5
# the green line approaching 1 as number of steps increases matches the mathematical conclusion 
```
